Detrás de cada aplicación hay un grupo de trabajo de desarrollo, las aplicaciones que no llegan a ser conocidas en el inmenso mercado de apps son una pérdida de esfuerzo e inlcluso en algunos casos pérdida monetaria para el grupo desarrollador. Este estudio permitirá a desarrolladores y empresas tener en cuenta algunas variables para hacer que su aplicación llegue a los resultados esperados con mayor probabilidad.

Para ello hemos escogido una data base con más de diez mil aplicaciones del mercado de Google Play, algunas variables son:

- Categoría: Las categorías ayudan a los usuarios a buscar y descubrir las apps más relevantes en Play Store, entre ellas están: Arte y diseño, autos y vehículos, belleza, libros y referncias, empresa, cómics, comunicaciones, citas, educacion, entretenimiento, eventos, finanza, comida y bebida, salud y bienestar, casa y hogar, bibliotecas y demstración, estilo de vida, mapas y navegación, medicina, música y audio, noticias y revistas, paternidad y maternidad, personalización, fotografía, productividad, compras, sociales, deportes, herramientas, viajes y local, reproductores y editores de videos y clima; Para videojuegos existen las siguientes categorías: Acción, Aventura, Carreras, Cartas, Casino, Casual, Deportes, Educativos, Estrategia, Juegos de mesa, Juegos de palabras, Juegos de rol, Música, Preguntas y respuestas, Rompecabezas, Sala de juegos, Simulación.

- Rating: En una escala de 1 a 5 estrellas los usuarios califican la aplicación según su experiencia con ella.
De esta variable se espera que las aplicaciones con mejor calificación sean las más descargadas por los usuarios, pues hay estudios en los que se muestra que una buena calificación le da valor por parte del usuario a la aplicación.

- Tamaño: Peso de la aplicación en MB (Megabytes). - Precio:Precio en dólares que debe pagar el usuario al descargar la aplicación. De la cual se espera que las aplicaciones con menor precio sean las que mayor descargas tienen.

>>-Maduro: Contenido dirigido a mayores de 17 años, puede incluir manifestaciones de violencia intensa, escenas sangrientas, contenido sexual o lenguaje malsonante.

>>-Adultos únicamente: Contenido dirigido a mayores de 18 años, puede contener escenas prolongadas de violencia intensa, contenido sexua explícito o juegos de azar con dinero real.

## **Calificación**

Veamos cómo se comportan todas estas aplicaciones en promedio. Las calificaciones son un indicador clave del rendimiento de una aplicación, por ende, se hará un histograma de las aplicaciones con respecto a las calificaciones, con el objetivo de ver su distribución y a partir de allí plantear la herramienta que mejor se pueda ajustar al análisis.


```python
fig_h = go.Figure(go.Histogram(
                  x = df_Apps['Rating'],
                  opacity = .5,
                  )
)
fig_h.show(renderer ='colab')
```

En primera instancia se observa que la mayoría de las aplicaciones tienen calificaciones altas, con solo unas pocas excepciones en las aplicaciones con calificaciones bajas. El gráfico del histograma está sesgado hacia la izquierda, es decir, el indice de disperción para la variable "Calificación" no es normal y en tal situación es mejor utilizar los cuartiles para un análisis más preciso. 

Los cuartiles nos brindan información en cuanto a la distribución de los datos, su disperción y su tendencia central. El cuartil es cada uno de los tres valores que pueden dividir un grupo de números, ordenados de menor a mayor, en cuatro partes iguales, cada cuartil determina la separación entre uno y otro subgrupo, dentro de un conjunto de valores estudiados. El primer cuartil (**Q1**) divide los valores en una relación de **25:75**, es decir, aquellos datos menores a Q1 representan el 25% de los datos, y los que son mayores a Q1 representan el 75% de los datos, así mismo, el segundo cuartil (**Q2**) divide los datos en una relación **50:50** (la conocemos como la mediana) y el tercer cuartil (**Q3**) divide los datos en una relación **75:25**.

El rango intercuartilico (**RIC**) es una medida de dispersión de un conjunto de datos que indica la distancia entre el primer cuartil y el tercer cuartil.  Utiliza la mediana como medida central, su representación grafica viene siendo el diagrama de caja y abarca el **50%** central de los datos.


```python
#Primero filtramos las filas en que Rating son no nulos
filtro = df_Apps[~(df_Apps['Rating'].isnull())]

traza = go.Box(
    x = df_Apps['Type'],
    y = df_Apps['Rating'],
    name = 'Rating'
)

data = [traza]
fig = go.Figure(data = data)
fig.show(renderer ='colab')
```

El diagrama de cajas nos sirve para ver el centro, la dispersión y la asimetría de un conjunto de datos.
 

Como podemos observar el **50%** de las aplicaciones gratuitas tienen una calificación entre **4.0** y **4.5** mientras que las aplicaciones que son pagas estan entre **4.1** y **4.6** aproximadamente, no es mucha la diferencia, pero si podríamos afirmar que las aplicaciones pagas son mejor calificadas. Ambos diagramas son muy similares, por ejemplo, tanto las aplicaciones pagas como las gratuitas tienen calificaciones sesgadas al extremo superior porque el **50%** inferior de los valores se distribuyen en un rango más grande que el **50%** superior de los valores.

Google play mide el tamaño de las aplicaciones en MB (Megabites), es una unidad de información equivalente a un millón de Bytes.

```python
#Haremos una refgresion logistica sobre la variable Type
df_Apps["Type"].replace({'Free':'1','Paid':'0'}, regex=True, inplace = True)
df_Apps["Installs"].replace({'\+':'','\,':''}, regex=True, inplace = True)
df_Apps = df_Apps.astype({'Type': 'uint8','Installs': 'float'})
```

```python
#Ajuste del Modelo regresion simple
X = df_Apps[['Installs']]
y = df_Apps['Type']
X_train, X_test, y_train, y_test = train_test_split(
                                        X.values.reshape(-1,1),
                                        y.values.reshape(-1,1),
                                        train_size   = 0.7,
                                        random_state = 0,
                                        shuffle      = True
                                    )
modelo = LogisticRegression(penalty='none')
modelo.fit(X = X_train.reshape(-1, 1), y = y_train)
```
